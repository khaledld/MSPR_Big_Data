```python
import numpy as np
import pandas as pd
  
emp_data = pd.read_excel("E:/Ingénierie 1 EISI/MSPR 1/Projet MSPR Big Data/Datasets/Used Datasets/base-cc-emploi-pop-active-2020_v2.xlsx", sheet_name='COM_2020')
emp_data
```

```python
new_col = emp_data.iloc[4]
emp_data.columns = new_col
emp_data
```

```python

new_data = emp_data.drop(emp_data.index[:5])
new_data.reset_index(drop=True)
new_data
```

```python
new_data.columns.name = ""
new_data = new_data.reset_index(drop=True)
new_data
```

```python
needed_cols_1 = ['CODGEO', 'P20_POP1564', 'P20_ACT1564', 'P20_CHOM1564', 'P20_ACTOCC1564', 'C20_EMPLT', 'C20_EMPLT_CS6']
```

```python
data_1 = new_data
```

```python
not_needed_col_1 = [colonne for colonne in data_1.columns if colonne not in needed_cols_1]
```

```python
data_1 = data_1.drop(not_needed_col_1, axis=1)
```

```python
data_1
```

```python
new_df_cols = ["Code géographique", "Taux d'activité des 15-64 ans", "Taux d'emploi des 15-64 ans", "Taux de chômage des 15-64 ans", "Part des ouvriers dans les emplois"]
new_df = pd.DataFrame(columns=new_df_cols)
```

```python
new_df
```

```python
new_df['Code géographique'] = data_1['CODGEO']
```

```python
data_1['P20_POP1564'] = data_1['P20_POP1564'].replace(0, pd.NA)
```

```python
new_df['Taux d\'activité des 15-64 ans'] = 100 * data_1['P20_ACT1564'] / data_1['P20_POP1564']
new_df
```

```python

new_df['Taux d\'emploi des 15-64 ans'] = 100 * data_1['P20_ACTOCC1564'] / data_1['P20_POP1564']
new_df
```

```python
new_df['Taux de chômage des 15-64 ans'] = 100 * data_1['P20_CHOM1564'] / data_1['P20_POP1564']
new_df
```

```python
data_1['C20_EMPLT'] = data_1['C20_EMPLT'].replace(0, pd.NA)
new_df['Part des ouvriers dans les emplois'] = 100 * data_1['C20_EMPLT_CS6'] / data_1['C20_EMPLT']
new_df
```

```python
import os

rep_dir = "E:\Ingénierie 1 EISI\MSPR 1\Projet MSPR Big Data\Datasets\Extraxted Data"
if not os.path.exists(rep_dir):
    os.makedirs(rep_dir)
csv_file_dir = os.path.join(rep_dir, "donnees_emploi.csv")
new_df.to_csv(csv_file_dir, index=False)
```
