```python
#Classification des candidats selon leur orientation politique
# "EXG" pour extrême gauche
# "EXD" pour extrême droite
# "G" pour gauche
# "D" pour droite
# "c" pour centre


import numpy as np


conditions = [
    (df_elus['Elu'].isin(['Nathalie', 'Jean-Luc', 'Philippe'])),  # Condition 1
    (df_elus['Elu'].isin(['Anne', 'Yannick', 'Fabien'])),  # Condition 2
    (df_elus['Elu'].isin(['Emmanuel', 'Jean'])),  # Condition 3
    (df_elus['Elu'].isin(['Valérie', 'Nicolas'])),  # Condition 3
    (df_elus['Elu'].isin(['Éric', 'Marine']))   # Condition 5
]

df_elus['Orientation'] = np.select(conditions, ['EXG', 'G', 'C', 'D', 'EXD'], default='')


df_elus
```
