```python
import pandas as pd

delin = pd.read_csv("E:/Ingénierie 1 EISI/MSPR 1/Projet MSPR Big Data/Datasets/Extraxted Data/délinquence_17_22.csv")
empl = pd.read_csv("E:/Ingénierie 1 EISI/MSPR 1/Projet MSPR Big Data/Datasets/Extraxted Data/donnees_emploi.csv")
med_niv_vie = pd.read_csv("E:/Ingénierie 1 EISI/MSPR 1/Projet MSPR Big Data/Datasets/Extraxted Data/mediane_niveau_vie.csv")
taux_p = pd.read_csv("E:/Ingénierie 1 EISI/MSPR 1/Projet MSPR Big Data/Datasets/Extraxted Data/taux_pauvrete.csv")
part_elu = pd.read_csv("E:/Ingénierie 1 EISI/MSPR 1/Projet MSPR Big Data/Datasets/Extraxted Data/PartiElu.csv")
part_elu = pd.read_csv("E:/Ingénierie 1 EISI/MSPR 1/Projet MSPR Big Data/Datasets/Extraxted Data/PartiEluEX.csv")
```

```python
df = pd.merge(med_niv_vie, empl, on="Code géographique")
```

```python
df = pd.merge(df, delin, on="Code géographique")
df = pd.merge(df, taux_p, on="Code géographique")
df = pd.merge(df, part_elu, on="Code géographique")
```
